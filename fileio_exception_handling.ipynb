```python
###########
# File IO # 
###########

############ Reading a file ############

f = open("file_io_read.txt", "r") #read only mode
print(f.read(), "\n----------") #prints entire text file
f.close() #otherwise maybe memory leak

#context manager - opens and closes file automatically
with open("file_io_read.txt", "r") as f: #each line read only once, whatever already read, not read again
    # print(f.read()) #prints everthing
    # print(f.read(100)) #reads first 100 chars, if file < 100 chars, no problem
    print(f.readline()) #prints first line, ends in \n
    print(f.readlines(), "\n------") #all lines read as list, first line already read so not read again, turn off above and see

with open("file_io_read.txt", "r") as f: #where is the read pointer
    print(f.read(10))
    print(f.read(20))
    print(f.read(5))
    print(f.tell(), "\n----") #read pointer at 35

with open("file_io_read.txt", "r") as file:
    lines = file.read().splitlines()  # Read the lines of the file and split it into list elements
print(lines, "\n--------")

with open("file_io_read.txt", "r") as f: #using a for loop, sometimes helpful
    for line in f:
        print(line, end="") #default end is \n so will print \n after each line, use end="" to avoid this
    print("\n------")

with open("download.png", "rb") as f: #read a binary format file
    binary_file = f.read()
    print(binary_file)

############ Writing to a file ############

with open("file_io_write.txt", "w") as f: #creates a new file and writes to it. If exists, rewrites all data
    f.write("A new file created and some text has been written")

with open("write_binary.png", "wb") as f: #use 'wb' to write in a binary format
    f.write(binary_file)

#Note : use 'a' flag to append to an existing file, If the file does not exist, it creates a new file for writing.
# see https://www.tutorialspoint.com/python/python_files_io.htm for all flags
```

```python
######################
# Exception Handling #
######################
try:
    #x = 10*(1/0)
    x = 10+'10'
    print(x)
except ZeroDivisionError as e:
    print(f'Exception : {e}')
except TypeError as e:
    print(f'Type error Exception : {e}')
except Exception as e: #catch-all block, putting this on top catches all exceptions making other blocks unreachable
    print(f'Generic Exception : {e}')
finally :
    print("program end")

try:
    x = None
    if x is None:
        raise Exception("Uh oh") #raising an exception for some condition
except ZeroDivisionError as e:
    print(f'Exception : {e}')
except Exception as e:
    print(f'Generic Exception : {e}')
```
