```python
###########
# Strings #
###########
#triple quotes usage
triple_quotes_exmaple = '''You are
so beautiful to me
cant you see?'''
print(triple_quotes_exmaple)

#formatting options and concatenation
first_name = "Sid"
last_name = "Yash"
#loose coupling, leads to bugs
format_text = "Hello, {} {}".format(first_name, last_name)
print(format_text)
#tight coupling
format_text = "Hello tightly coupled, {first} {last}".format(first=first_name, last=last_name)
print(format_text)
# ints and strings
number = 686
format_text = f"Hello {number}"
print(format_text)

#check for digits in a string
var = '100'
print("Is digit?", var.isdigit())

#user input
name = input("Enter your name: ") #dialog opens even tho assigned to variable, nothing to do with print below
print(f"My name is : {name}")

#indexing, slicing, split
greeting = "Welcome, to the black parade"
print(greeting[2]) #indexing
print(greeting[2:10:2]) #slicing, ignores whitespace?
print(len(greeting)) #includes whitespace
print(greeting.split())
print(greeting.split(','))#split by comma
print(greeting.find("to"))#return start index of 'to'

#joins method
print(":".join('abcdefg')) #input string, inserts : between all literals of string
print(":".join(('ab', 'cd', 'ef','g'))) #input tuple of strings, joins each element of tuple

name = ' sid yash yash '
print(name.strip()) #strips away whitespaces at begin and end
print(name.replace('yas', 'bus')) #replaces all occurences of string yas
```

```python
###########
# Lists   #
###########
my_list = [1,2,3,4,5]
print("last element:", my_list[-1]) #prints last element
print("index of 4:", my_list.index(4)) #returns index of a value

#slicing
print(my_list[0:10:2]) #no index out of bounds, step size 1 -every element, 2- every 2nd element ...
print(my_list[1:3]) #excludes element at index 3

new_list = my_list[1:3]
print(new_list, my_list) #doesn't point to same memory, noice

my_list.append(12) #to end of list
my_list.insert(3,12) #insert 12 at pos 3
print(my_list)
my_list.remove(12) #removes first occurence of 12
print(my_list) 
my_list.pop() #removes last element, no popleft method :(
print(my_list) 
my_list.pop(0) #removes at index 0
print(my_list)
my_list.pop(-1) #removes the last element
print(my_list)
my_list.reverse()
print(my_list)
my_list.sort()
print(my_list)

#nested lists
nlist = [[1,2],[1,3,7],[3,8,7,4]]
print(nlist[-1])
print(nlist[1][-1])

#memory reference - copy method
list1 = [1,2,3,4,5]
list2 = list1
list2.pop()
print(list1, list2) #changing assignment variable still points to same variable
list1 = [1,2,3,4,5]
list2 = list1.copy() #creates a new list in memory
list2.pop()
print(list1, list2)
```

```python
#############
# Functions #
#############
def func1(*args): #pass in multiple args
    print(args[-1])
func1(1,'a',2,3,'f') # not a list rather arguments

def fib(seq): #recursive fibonacci exmaple
    #print(seq)
    seq.append(seq[len(seq)-1]+seq[len(seq)-2])
    #print("--------")
    if len(seq) == 10:
        return seq
    else:
       seq = fib(seq)
    return seq
seq = fib([1,1])
print(seq)

def fib_pos(pos): #recursive fibonacci exmaple - for one pos
    if pos == 0:
        return 1
    elif pos == 1:
        return 1
    else:
        return fib_pos(pos-2)+fib_pos(pos-1)

for x in range(10):
    print(fib_pos(x))

def factorial(x,y): #recursive factorial example BAD
    x = x*(y-1)
    y -= 1
    if y == 2:
        return x
    else:
        x = factorial(x,y)
    return x
print(factorial(5,5))

def factorial_good(x): #recursive factorial example GOOD
    if x == 1:
        return 1
    else:
        return x * factorial_good(x-1)
print(factorial_good(5))
```

```python
################
# Dictionaries #
################
#keys must always be an immutable data type, such as strings, numbers, or tuples
#values can be any data types
dictionary = {
  1: 'hello', 
  'two': True, 
  '3': [1, 2, 3], 
  'Four': {'fun': 'addition'}, 
  5.0: 5.5
}
print(dictionary[1], dictionary['two']) #access by exact key value
# dictionary.get(keyname, value) - A value to return if the specified key does not exist. Default value None
print(dictionary.get(1), dictionary.get('two'), dictionary.get('three', -1)) #same but with get() method
dictionary['major'] = 'Math' #adding a new key-val pair
print(dictionary)
print(len(dictionary)) #find no of key val pairs

#pop methods
dictionary.pop(1) #needs a key as an anrgument 
dictionary.popitem() #pops last element, needs no argument
print(dictionary)

#delete methods
del dictionary['two']
print(dictionary)
#del dictionary # deletes entire dictionary
dictionary.clear() # removes all key val pairs, but doesnt del dictionary
print(dictionary)

#Updating dictionaries with the .update() method
dict1 = {'color': 'blue', 'shape': 'circle'}
dict2 = {'color': 'red', 'number': 42}
dict1.update(dict2)
print("dict1: ", dict1)

#working with keys and values
#using loops
for x in dict1: print(x)#prints keys as it is
for x,y in dict1.items(): print(x,y) #prints keys and values as it is
    
if 'color' in dict1: print("key exists") #check if key exists
# remember the second param value of get() method can avoid this 'if' in most cases
print(dict1.get('texture', 'sorry not found'))

#in-built methods
print(dict1.keys()) #returns a list of keys
print(dict1.values()) #returns a list of vals
print(dict1.items()) #retrns keys-vals as a list of tuples

#copy method-bcz reassigning to new variable doesnt change reference, just like lists
dict2 = dict1.copy()
dict2.popitem()
print(dict1, dict2)
```

```python
###########
# Tuples  #
###########
# A tuple is a collection which is ordered -  items have a defined order, and that order will not change.
#  and unchangeable - we cannot change, add or remove items after the tuple has been created.

mytuple = ("apple", "banana", 34, True, 40, "male") #allow duplicate values, diff data types
print(mytuple[2], mytuple[-1]) #indexing
print(len(mytuple))
print(mytuple[1:4]) #slicing
# mytuple[2] = "nice" - gives an error, cannot change elements

one_item_tuple = ("apple",) #comma at end
thistuple = tuple(("apple", "banana", "cherry")) # constructor: note the double round-brackets
```

```python
###########
# Sets    # 
###########
#unique items only, unordered, unchangeable, can add/remove but not change items(no indexing changes)
myset = {"abc", 34, True, 40, "male"}  #allow diff data types
print(len(myset))
print(myset)

print("abc" in myset) #check if element exists in a set
# print(myset[3]) #unordered so gives an error

myset.add("apple") #add an item
print(myset)
myset.update(["mango", "pear"]) #add multiple items
print(myset)

myset.remove("male") #error if item does not exist
print(myset)
myset.discard("female") #same as remove but no error if item not present
print(myset)

myset2 = {1,2,34}
print("union", myset2.union(myset)) #notice how 1 and True are considered same, same for 0 and False
print(myset.intersection(myset2))
print(myset.difference(myset2))

myset.clear()
print(myset)
```
